```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
statistics = pd.read_csv(r'I:\Data\Personal Data\graduation project\Code\Machine Learning\basicdata\2007_statistics.csv',index_col='Unnamed: 0')
statistics.dropna(inplace=True)
```

```python
flag = 0
x_ticks = []
x_labels = []
for time in range(len(statistics.index)):
    if(flag != statistics.index[time][5:7]):
        flag = statistics.index[time][5:7]
        x_ticks.append(time)
        x_labels.append(flag)
# x_ticks.append(len(lv1_data['Date/Time'])-1)
# x_labels.append(lv1_data['Date/Time'][len(lv1_data['Date/Time'])][9:14])
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.figure(figsize=(12,7))
# plt.plot(statistics['Ground_temperature_08'],label='微波辐射计')
# plt.plot(statistics['Temperature_08_H0'],label='探空数据')
plt.plot(statistics['Ground_temperature_08']-statistics['Temperature_08_H0'],label='差异')
plt.legend()
plt.xticks(x_ticks,x_ticks)
plt.xlabel('样本数量',fontsize=22)
plt.ylabel('差异',fontsize=22)
plt.savefig(r'I:\Data\Personal Data\graduation project\毕业论文\毕业论文图集\Chapter5\2007_N_TkBB(K)_difference.svg',bbox_inches='tight',dpi=300,pad_inches=0)
plt.show()
```
